```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("dataset_44_spambase.csv")
```

```python
df.head()
```

```python
df.tail()
```

```python
#setting up our target variable
target=df.pop("class")
```

```python
target
```

```python
#setting up the features
features=df
features
```

```python
#initializing the model
rfmodel=RandomForestClassifier(random_state=1)
```

```python
#fitting the data into our model
rfmodel.fit(features,target)
```

```python
#predicting
preds=rfmodel.predict(features)
```

```python
#predicted values
preds
```

```python
#cheking the score- level of diff between predicted and actual target values
score=accuracy_score(preds,target)
```

```python
score
#we have achieved an accuracy of 99.65%
```
